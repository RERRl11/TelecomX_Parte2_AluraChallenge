#üõ†Ô∏è Preparaci√≥n de los Datos

##Extracci√≥n del Archivo Tratado

```python
import pandas as pd

df = pd.read_csv("TelecomX_Data_Procesados.csv")

display(df.head())
```

```python
display(df.info())
```

```python
display(df.isnull().sum())
```

```python
display(df.describe())
```

##Eliminaci√≥n de Columnas Irrelevantes

```python
df = df.drop('ID_Cliente', axis=1)
display(df.head())
```

##Encoding

```python
df = pd.get_dummies(df, columns=['Servicio_Internet', 'Contrato', 'Metodo_Pago'])
display(df.head())
```

##Verificaci√≥n de la Proporci√≥n de Cancelaci√≥n (Churn)

```python
# Calcular la proporci√≥n de clientes que cancelaron (Evasion = 1) vs no cancelaron (Evasion = 0)
churn_counts = df['Evasion'].value_counts(normalize=True)

display(churn_counts)
```

##Balanceo de Clases

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Separar las caracter√≠sticas (X) y la variable objetivo (y)
X = df.drop('Evasion', axis=1)
y = df['Evasion']

# Mostrar la distribuci√≥n original de las clases
print('Distribuci√≥n original del conjunto de datos %s' % Counter(y))

# Aplicar SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Mostrar la distribuci√≥n de las clases despu√©s de aplicar SMOTE
print('Distribuci√≥n del conjunto de datos despu√©s de SMOTE %s' % Counter(y_res))

# Opcional: Crear un nuevo DataFrame balanceado
df_balanced = pd.concat([X_res, y_res], axis=1)
display(df_balanced.head())
```

## üìä Divisi√≥n de Datos y Modelado

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Dividir los datos balanceados en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.25, random_state=42, stratify=y_res)

# Inicializar y entrenar el modelo de Regresi√≥n Log√≠stica
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluar el modelo
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

##Separaci√≥n de Datos

```python
from sklearn.model_selection import train_test_split

# Dividir los datos balanceados en conjuntos de entrenamiento y prueba (por ejemplo, 80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.20, random_state=42, stratify=y_res)

print("Forma del conjunto de entrenamiento (X_train):", X_train.shape)
print("Forma del conjunto de prueba (X_test):", X_test.shape)
print("Forma de las etiquetas de entrenamiento (y_train):", y_train.shape)
print("Forma de las etiquetas de prueba (y_test):", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# --- Modelo de Regresi√≥n Log√≠stica ---
print("--- Regresi√≥n Log√≠stica ---")
# Ya tenemos X_train, X_test, y_train, y_test de la divisi√≥n anterior usando datos balanceados y escalados

model_lr = LogisticRegression(random_state=42)
model_lr.fit(X_train, y_train)
y_pred_lr = model_lr.predict(X_test)

print("Matriz de Confusi√≥n (Regresi√≥n Log√≠stica):")
print(confusion_matrix(y_test, y_pred_lr))
print("\nReporte de Clasificaci√≥n (Regresi√≥n Log√≠stica):")
print(classification_report(y_test, y_pred_lr))

print("\n" + "="*50 + "\n") # Separador

# --- Modelo de Random Forest ---
print("--- Random Forest ---")

# Para Random Forest, podemos usar los datos balanceados (X_res, y_res) y la divisi√≥n (X_train, X_test, y_train, y_test)
# Random Forest no requiere escalado, pero usar los datos escalados no lo perjudica.

model_rf = RandomForestClassifier(random_state=42)
model_rf.fit(X_train, y_train) # Usamos los datos de entrenamiento balanceados y escalados
y_pred_rf = model_rf.predict(X_test) # Predecimos en el conjunto de prueba balanceado y escalado

print("Matriz de Confusi√≥n (Random Forest):")
print(confusion_matrix(y_test, y_pred_rf))
print("\nReporte de Clasificaci√≥n (Random Forest):")
print(classification_report(y_test, y_pred_rf))
```

## Importancia de la Normalizaci√≥n de Datos

La normalizaci√≥n de datos es un paso crucial en el preprocesamiento de datos para muchos algoritmos de machine learning. Su objetivo principal es asegurar que las caracter√≠sticas num√©ricas tengan una escala similar, evitando que aquellas con rangos de valores m√°s amplios dominen el proceso de modelado.

Esto es particularmente importante para modelos que se basan en:

1.  **Distancia:** Algoritmos como K-Nearest Neighbors (KNN) y Support Vector Machines (SVM) calculan distancias entre puntos de datos para hacer predicciones. Si las caracter√≠sticas no est√°n en la misma escala, la distancia estar√° dominada por la caracter√≠stica con el rango m√°s grande, sesgando los resultados. La normalizaci√≥n asegura que cada caracter√≠stica contribuya equitativamente al c√°lculo de la distancia.

2.  **Optimizaci√≥n de Par√°metros:** Modelos como la Regresi√≥n Log√≠stica y las Redes Neuronales a menudo utilizan algoritmos de optimizaci√≥n (como el descenso de gradiente) para encontrar los mejores par√°metros del modelo. Un espacio de caracter√≠sticas no escalado puede resultar en una superficie de funci√≥n de costo alargada y estrecha, lo que dificulta que el algoritmo de optimizaci√≥n converja de manera eficiente y r√°pida hacia el m√≠nimo global. La normalizaci√≥n crea una superficie de funci√≥n de costo m√°s sim√©trica, facilitando una convergencia m√°s r√°pida y estable.

En resumen, normalizar los datos ayuda a:

*   **Prevenir el sesgo por magnitud:** Evita que las caracter√≠sticas con valores m√°s grandes tengan un impacto desproporcionado.
*   **Mejorar la convergencia de los algoritmos:** Permite que los algoritmos de optimizaci√≥n encuentren soluciones de manera m√°s eficiente.
*   **Asegurar el correcto funcionamiento de modelos basados en distancia:** Garantiza que las distancias calculadas sean significativas.

Es por estas razones que, al trabajar con modelos sensibles a la escala, la normalizaci√≥n es una etapa fundamental en la preparaci√≥n de los datos.

## üìà An√°lisis Cr√≠tico y Comparaci√≥n de Modelos

Bas√°ndonos en los resultados de la evaluaci√≥n de los modelos de Regresi√≥n Log√≠stica y Random Forest en el conjunto de datos balanceado, podemos realizar un an√°lisis cr√≠tico y comparar su rendimiento:

**¬øCu√°l modelo tuvo el mejor desempe√±o?**

Seg√∫n las m√©tricas de evaluaci√≥n (Precisi√≥n, Recall, F1-score y Exactitud) en el conjunto de prueba, el modelo de **Random Forest** demostr√≥ un mejor desempe√±o general en comparaci√≥n con el modelo de Regresi√≥n Log√≠stica. Espec√≠ficamente para la clase de inter√©s (Evasi√≥n = 1.0), Random Forest obtuvo valores m√°s altos en Precisi√≥n (0.83 vs 0.79), Recall (0.88 vs 0.85) y F1-score (0.85 vs 0.82). Esto indica que el modelo de Random Forest es m√°s efectivo tanto para identificar correctamente a los clientes que cancelar√°n (mayor Recall) como para reducir la cantidad de falsos positivos (mayor Precisi√≥n).

**¬øAlg√∫n modelo present√≥ overfitting o underfitting?**

Para evaluar el overfitting o underfitting, necesitar√≠amos comparar el rendimiento de los modelos en el conjunto de entrenamiento y en el conjunto de prueba. Un rendimiento significativamente mejor en el conjunto de entrenamiento que en el de prueba podr√≠a indicar overfitting, mientras que un rendimiento bajo en ambos conjuntos podr√≠a sugerir underfitting.

En este an√°lisis inicial, no calculamos expl√≠citamente las m√©tricas en el conjunto de entrenamiento. Sin embargo, podemos inferir lo siguiente bas√°ndonos en los resultados en el conjunto de prueba balanceado:

*   **Regresi√≥n Log√≠stica:** El rendimiento general es razonable (Accuracy de 0.81). No parece haber un claro indicio de underfitting severo, ya que el modelo logra capturar algunas tendencias en los datos para predecir la evasi√≥n. Para confirmar overfitting, necesitar√≠amos comparar su rendimiento con el conjunto de entrenamiento. Dado que es un modelo lineal, es menos propenso al overfitting extremo en comparaci√≥n con modelos m√°s complejos si los datos no tienen patrones muy complejos.

*   **Random Forest:** Obtuvo un mejor rendimiento en el conjunto de prueba (Accuracy de 0.85). Los modelos de Random Forest, al ser modelos basados en √°rboles y ensambles, tienen un mayor potencial de overfitting si no se controlan sus hiperpar√°metros (como la profundidad de los √°rboles o el n√∫mero m√≠nimo de muestras por hoja). Para determinar si hay overfitting, ser√≠a crucial evaluar su rendimiento en el conjunto de entrenamiento y compararlo con el del conjunto de prueba. Si el rendimiento en entrenamiento es mucho mayor que en prueba, ser√≠a un indicio de overfitting.

**Posibles Causas y Ajustes:**

*   **Overfitting (si se observa en Random Forest):** Si el modelo de Random Forest muestra overfitting, las posibles causas podr√≠an ser que los √°rboles individuales se ajustan demasiado a los datos de entrenamiento. Los ajustes a considerar ser√≠an:
    *   Reducir la complejidad del modelo limitando la profundidad m√°xima de los √°rboles (`max_depth`).
    *   Aumentar el n√∫mero m√≠nimo de muestras requeridas para dividir un nodo (`min_samples_split`) o para que una hoja sea un nodo terminal (`min_samples_leaf`).
    *   Considerar la poda de √°rboles.
    *   Aumentar la cantidad de datos de entrenamiento (aunque ya aplicamos SMOTE para balancear).

*   **Underfitting (si se observa en cualquiera de los modelos):** Si un modelo presenta underfitting, significa que es demasiado simple para capturar la complejidad de las relaciones en los datos. Los ajustes a considerar ser√≠an:
    *   Para Regresi√≥n Log√≠stica: Explorar caracter√≠sticas adicionales, considerar interacciones entre caracter√≠sticas o probar modelos no lineales.
    *   Para Random Forest: Asegurarse de que los hiperpar√°metros no est√°n restringiendo excesivamente la complejidad del modelo (por ejemplo, `max_depth` demasiado bajo).
    *   Considerar la ingenier√≠a de caracter√≠sticas para crear variables m√°s informativas.
    *   Probar modelos m√°s complejos si es necesario.

Para una evaluaci√≥n definitiva de overfitting o underfitting, el pr√≥ximo paso l√≥gico ser√≠a calcular y comparar las m√©tricas de rendimiento en el conjunto de entrenamiento para ambos modelos.

## Resumen de los Datos y Preprocesamiento

El conjunto de datos inicial conten√≠a informaci√≥n sobre 7043 clientes, con 23 columnas que describen diversas caracter√≠sticas de los clientes y sus servicios, as√≠ como la variable objetivo `Evasion` (0 para no evasi√≥n, 1 para evasi√≥n). Los pasos iniciales de exploraci√≥n de datos revelaron que no hab√≠a valores faltantes en el conjunto de datos. Se examinaron las estad√≠sticas descriptivas para entender la distribuci√≥n de las variables num√©ricas y la proporci√≥n de las clases en la variable objetivo.

Como parte del preprocesamiento, se elimin√≥ la columna `ID_Cliente` ya que es un identificador √∫nico y no aporta informaci√≥n predictiva al modelo. Las variables categ√≥ricas `Servicio_Internet`, `Contrato` y `Metodo_Pago` fueron transformadas utilizando codificaci√≥n one-hot para convertirlas en un formato num√©rico adecuado para los algoritmos de machine learning.

La exploraci√≥n inicial de la variable objetivo `Evasion` mostr√≥ un desbalance de clases significativo, con una proporci√≥n mucho mayor de clientes que no evadieron (aproximadamente 73.5%) en comparaci√≥n con los que s√≠ evadieron (aproximadamente 26.5%). Para abordar este desbalance y mejorar el rendimiento del modelo en la predicci√≥n de la clase minoritaria (evasi√≥n), se aplic√≥ la t√©cnica de sobremuestreo SMOTE (Synthetic Minority Over-sampling Technique). Despu√©s de aplicar SMOTE, la distribuci√≥n de las clases se balance√≥, resultando en un n√∫mero igual de instancias para ambas clases (no evasi√≥n y evasi√≥n).

Finalmente, los datos preprocesados y balanceados fueron divididos en conjuntos de entrenamiento y prueba para preparar los datos para el modelado predictivo.

Al dirigir estas estrategias a los segmentos de clientes identificados como de alto riesgo y abordar los factores espec√≠ficos que impulsan la evasi√≥n, TelecomX puede mejorar significativamente sus tasas de retenci√≥n de clientes.
